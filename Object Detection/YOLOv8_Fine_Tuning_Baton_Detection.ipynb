```python
import cv2
import os
import urllib.request
from PIL import Image
from ultralytics import YOLO
import torch
import matplotlib.pyplot as plt
from tqdm import tqdm
```

## Fine-Tuning (Training)

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(data="/content/drive/MyDrive/Spark Project/Data (For YOLOv8 Training)/Option 3 - Roboflow (60 Images)/data.yaml", epochs = 200)
```

## Inference (Post Training)

```python
model = YOLO("/content/drive/MyDrive/Spark Project/Data (For YOLOv8 Training)/Option 3 - Roboflow (60 Images)/YOLOv8 Best Weights.pt")
```

```python
test_images = os.listdir("/content/drive/MyDrive/Spark Project/Data (For YOLOv8 Training)/Option 3 - Roboflow (60 Images)/test/")
for i in test_images:
  results = model(os.path.join("test_images", i))
  for r in results:
    im_array = r.plot()
    img = Image.fromarray(im_array[..., ::-1])
    img.save(os.path.join("/content/drive/MyDrive/Spark Project/Data (For YOLOv8 Training)/Option 3 - Roboflow (60 Images)/prediction plots", i))
#results = model("frame64728.jpg", mode = "detect")
```
